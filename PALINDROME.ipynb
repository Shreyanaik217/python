```python
lower=int(input('enter the lower value:'))
upper=int(input('enter the upper  value:'))
print('the palindrome number from',lower,'and',upper,'are')
for n in range(lower,upper+1):
    temp=n
    Rev=0
    while(temp>0):
        Rem=temp%10
        Rev=(Rev*10)+Rem
        temp=temp//10
    if(n==Rev):
        print(n)
    
```
